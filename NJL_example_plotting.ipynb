# Plotting macro for NJL paper #

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize 
from IPython.display import display, Math, Latex
```

```python
#GATHERING DATA FROM THE PLOTS

BR_3S_Babar = np.genfromtxt("LQpairproduction.dat")
BR_1S_Babar = np.genfromtxt("LQpairproduction.dat")
BR_1S_Belle = np.genfromtxt("LQpairproduction.dat")
BR_1S_CLEO = np.genfromtxt("LQpairproduction.dat")
```

```python
#MASSES AND cross-sections

mass_3S_Babar = BR_3S_Babar[:,0]
brunch_3S_Babar = BR_3S_Babar[:,1]

mass_1S_Babar = BR_1S_Babar[:,0]
brunch_1S_Babar = BR_1S_Babar[:,1]

mass_1S_Belle = BR_1S_Belle[:,0]
brunch_1S_Belle = BR_1S_Belle[:,1]

mass_1S_CLEO = BR_1S_CLEO[:,0]
brunch_1S_CLEO = BR_1S_CLEO[:,1]
```

```python
#PLOT CHECK 

plt.axis([0,9,10**-18,1])
plt.yscale('log')
plt.xlabel('$m_a$ [GeV]')
plt.title('Brunching Ratio x $10^{6}$')
plt.plot(mass_3S_Babar, brunch_3S_Babar, label = 'Babar 3S')
plt.plot(mass_1S_Babar,brunch_1S_Babar, label = 'Babar 1S')
plt.plot(mass_1S_Belle, brunch_1S_Belle, label = 'Belle 1S')
plt.plot(mass_1S_CLEO, brunch_1S_CLEO,label = 'CLEO')
plt.legend(loc='upper left')
plt.show()
```
